# 1) История

## 1.1) Кто придумал Python?
Python был создан **Гвидо ван Россумом** ([Guido van Rossum](https://ru.wikipedia.org/wiki/%D0%A0%D0%BE%D1%81%D1%81%D1%83%D0%BC,_%D0%93%D0%B2%D0%B8%D0%B4%D0%BE_%D0%B2%D0%B0%D0%BD)) в конце 1980-х годов, а первая версия вышла в 1991 году.

Гвидо — программист из Нидерландов. Он работал в Голландском центре математики и информатики (CWI, Amsterdam), и в конце 80-х решил создать свой язык программирования, вдохновлённый опытом работы с языком ABC (учебный язык, похожий на Pascal, но проще). <br>
image.png

## 1.2) Зачем он это сделал?
Гвидо хотел создать язык, который был бы:

*    понятным и читаемым для человека;
*    мощным, но не перегруженным синтаксисом;
*    удобным как для начинающих, так и для профессионалов;
*    легко расширяемым и пригодным для автоматизации повседневных задач.

По сути, он создавал язык, которым сам хотел пользоваться каждый день.


## 1.3) 🐍 Почему язык называется Python?
Итак, при чём тут змеи, а если быть точным -- **питон**?

Ответ: **не при чём**.
Название Python происходит от британского комедийного шоу [“Monty Python’s Flying Circus”](https://ru.wikipedia.org/wiki/%D0%9B%D0%B5%D1%82%D0%B0%D1%8E%D1%89%D0%B8%D0%B9_%D1%86%D0%B8%D1%80%D0%BA_%D0%9C%D0%BE%D0%BD%D1%82%D0%B8_%D0%9F%D0%B0%D0%B9%D1%82%D0%BE%D0%BD%D0%B0).

Гвидо был большим фанатом этого абсурдного и умного юмора и хотел, чтобы язык ассоциировался не с чем-то скучным и академичным, а с лёгкостью, игрой, творчеством.

# 3) Как писать код на Пайтоне

## 3.1) Синтаксис

📖 **Синтаксис** — это «грамматика» языка программирования.
Как в любом естественном языке у нас есть правила (где ставить запятые, как строить предложения), так и в Пайтоне есть строгие правила написания кода. В целом синтаксис Пайтона по возможности приближен к синтаксису европейский языков. Например, пробелы разделяют слова друг от друга, а круглые скобки определяют порядок арифметических операций как в математике.

Вообще правил в Пайтоне не мало, поэтому запоминать их будем по мере поступления.

Также важной частью синтаксиса является <font color="#6055ff"><b>PEP 8</b></font>. Коротко рассмотрим, что этот такое

## 3.2) PEP 8

🧾 Что такое <font color="#6055ff"><b>PEP 8</b></font>?

<font color="#6055ff"><b>PEP 8</b></font> — это официальное руководство по стилю написания кода на Python.
PEP расшифровывается как **Python Enhancement Proposal** — предложение по улучшению Python.

📌 А <font color="#6055ff"><b>PEP 8</b></font> — это документ, в котором описано, как должен выглядеть “красивый”, читаемый, питоновский код.

🎯 Зачем нужен <font color="#6055ff"><b>PEP 8</b></font>?
Python славится своей читаемостью и простотой, а <font color="#6055ff"><b>PEP 8</b></font> помогает:

*    писать код, который легко читать другим людьми;

*    соблюдать единые правила оформления в команде;

*    избегать хаоса, когда у каждого свой стиль написания.

Приведу два примера кода:

```python
# пример №1
2+2
```

```python
# пример №2
2 + 2
```

Видно, что результат выполнения кода одинаков, но если в первом примере чи́сла и знак `+` пишутся слитно, то во втором вокруг знак `+` стоит по одному знаку пробела. Оба варианта с точки зрения синтаксиса рабочие, но по правилам <font color="#6055ff"><b>PEP 8</b></font> второй вариант является предпочтительным. <font color="#6055ff"><b>PEP 8</b></font> *рекомендует ставить один пробел с каждой стороны арифметического оператора*. Дословно правило звучит следующим образом и его можно прочитать на [сайте](https://peps.python.org/pep-0008/) (если интересно):

>    Avoid trailing whitespace anywhere. Because it’s usually invisible, it can be confusing: e.g. a backslash followed by a space and a newline does not count as a line continuation marker. Some editors don’t preserve it and many projects (like CPython itself) have pre-commit hooks that reject it.

>    Always surround these binary operators with a single space on either side: assignment (=), augmented assignment (+=, -= etc.), comparisons (==, <, >, !=, <=, >=, in, not in, is, is not), Booleans (and, or, not).

>    If operators with different priorities are used, consider adding whitespace around the operators with the lowest priority(ies). Use your own judgment; however, never use more than one space, and always have the same amount of whitespace on both sides of a binary operator.

Вообще в данном курсе требования следовать стандартам <font color="#6055ff"><b>PEP 8</b></font> нет, поэтому *можете писать как желаете, лишь бы работало*. Тем не менее сделаю два небольших замечания:

*    При трудоустройстве работодатель может обращать внимание на то, как соискатель оформляет код, поэтому соблюдение стандартам может повлиять на поиск работы;

*    По своему опыту скажу, что на правильно оформленный код реально гораздо приятнее смотреть, не говоря уже о лучшей читаемости и понимании написанного

Так что я бы посоветовал сразу учиться писать код в соответствии стандартам 😊. Со своей стороны я буду отмечать то, как писать код можно в соответствии с <font color="#aa5500"><b>синтаксисом</b></font>, и как правильнее по стандартам <font color="#6055ff"><b>PEP 8</b></font>.

# 4) Операторы

Ну вот, уже почти подошли к практике. Поговорим об операторах.


> ➗ Оператор — это символ или слово, с помощью которого выполняется операция над данными.
Фактически большая часть операторов являются математическими знаками, а выполняемая ими операция известна нам из той же математики.

В Пайтоне есть следующие операторы, представленные знаками (то есть не словами из букв):

*    **Арифметические**: +, -, \*, /, //, %, **

*    **Сравнения**: ==, !=, <, >, <=, >=

*    **Логические**: &, |, not

*    **Присваивания**: =, +=, -=, *=, и т.д.

Посмотрим, как они работают


## 4.1) Арифметические операторы

```python
# + выполняет операцию сложения
2 + 2
```

```python
# - выполняет операцию вычитания
10 - 1
```

```python
# * выполняет операцию умножения
2 * 3
```

```python
# / выполняет операцию деления
9 / 3
```

Следующие операторы также выполняют математические операции. Они отличаются от стандартной математической записи в силу невозможности использовать привычную форму записи на компьютере:

```python
# // выполняет операцию целочисленного деления
5 // 2
```

Замечу, что по правилам <font color="#aa5500"><b>синтаксиса</b></font> операторы, состоящие из двух знаков пишутся слитно без пробелов. В противном случае Пайтон сообщит ошибке:

```python
4 / / 2
```

```python
# % возвращает остаток от целочисленного деления
10 % 3
```

```python
# ** выполняет операцию возведения в степень
10**3
```

```python
# Также как и в математике степень числа может быть дробною и отрицательною.
# Результат выполнения операции будет соответствующим
# Например, возведение в 0 любого числа вернёт единицу

123**0
```

```python
# Дробные степени от 0 до 1 невключительно являются извлечением корня
16**(1/2)
```

```python
# отрицательные числа "переворачивают" дробь
(1/2)**-1
```

```python
10**-1
```

## 4.2) Операторы сравнения

```python
# > -- это знак больше
3 > 2
```

```python
# < -- это знак меньше
3 < 2
```

Здесь заметим, что если результатом работы арифметического оператора является число, то операторы сравнения возвращают логическое значение. **True** -- если неравенство истинно, и **False** -- если ложно

```python
# >= -- это знак "больше либо равно" ≥
12 >= 12
```

```python
# >= -- это знак "меньше либо равно" ≤
100 <= 10
```

```python
# == выполняет функцию обычного знака равенства "=" как в математике (тождественно равно)
2 + 2 == 4
```

Почему тождественно равно == это удвоенный знак равенства, а не одинарный -- поясню позже

```python
# != знак неравенства
2 * 2 != 4
```

На примере операторов сравнения уже можно заметить расхождение с алгеброй. Если в математике знаки равенства или неравенства исходят только из истины (то есть выражение 2+2=5 ошибочно и его нельзя записать), то в Пайтоне оператор возвращает логический результат сравнения.

## 4.3) Логические операторы

```python
# & -- логическое «И» также известное как конъюнкция, логическое умножение
True & True
```

```python
# если хотя бы одно выражение ложно -- результат будет ложным
True & False & True
```

```python
# | -- логическое «ИЛИ» также известное как дизъюнкция, логическое сложение
False | False | False | False | True
```

```python
# not Логическое «НЕ» также называется отрицанием, инверсией
not True
```

И вот оператор `not` становится нашим первым словесным оператором. Кроме того, операторы `&` и `|` также имеют словесный аналог:


```python
True and False and True
```

```python
False or True or False
```

Пользоваться можно любым вариантом

## 4.4) Операторы присвоения

Итак, перейдём к группе операторов, которым нет аналога в математике. Что означает следующее алгебраическое выражение?

$$x^2 - x - 12 = 0$$
$$
\left[
\begin{aligned}
x &= -3 \\
x &= 4
\end{aligned}
\right.
$$

x -- теоретически это абсолютно любое число, но чтобы приведённое равенство было верным х должен быть равным либо -3, либо 4.
В Пайтоне знак равенства означает, что мы сохраняем в память машины значение справа от знака `=` под именем, написанным слева от знака `=`. Как это работает:


```python
x**2 - x - 12 == 0
```

Видимо ошибку `NameError: name 'x' is not defined`. Пайтон сообщает, что в его памяти нет такой переменной `x`. Пайтон не может работать с математическими абстракциями, как это может делать человек. Попробуем добавить в память устрайство переменную `x`:

```python
# Запишем в память Пайтона следующее и запустим ячейку
x = 1
```

```python
# Проверим равенство
x**2 - x - 12 == 0
```

Оператор `==` вернул значение `False`. Пайтон извлёк из своей памяти значение переменной `x` и использовал именно его при выполнении операций `**` (возведение в степень), `-` (вычитание), `==` (тождественно равно).
При выполнении операций получаем `11`. Так `11` не равно `0`, Пайтон возвращает логическое `False`. Теперь попробуем подставить `-3` и `4`:

```python
x = -3
x**2 - x - 12 == 0
```

```python
x = 4
x**2 - x - 12 == 0
```

Теперь всё верно.
Итак, в чём разница между `=` в математике и в программировании?

| Область       | Запись  | Значение                                  |
| ------------- | ------- | ----------------------------------------- |
| 🧮 Математика | `x = 5` | x всегда равен 5 (заявление факта)        |
| 🐍 Python     | `x = 5` | сохранить 5 в переменную x (присваивание) |

🔸 $$Математика: x = x + 1 — невозможно$$

🔸 Python: `x = x + 1` — абсолютно нормально! Увеличиваем значение `x` на 1 и сохраняем обратно.

➕ Расширенные операторы присваивания

Эти операторы сочетают арифметику и присваивание. Они эквивалентны длинной форме:

| Оператор | Что делает            | Пример    | То же самое, что... |
| -------- | --------------------- | --------- | ------------------- |
| `+=`     | прибавление           | `x += 1`  | `x = x + 1`         |
| `-=`     | вычитание             | `x -= 2`  | `x = x - 2`         |
| `*=`     | умножение             | `x *= 3`  | `x = x * 3`         |
| `/=`     | деление               | `x /= 4`  | `x = x / 4`         |
| `//=`    | целочисленное деление | `x //= 5` | `x = x // 5`        |
| `%=`     | остаток от деления    | `x %= 6`  | `x = x % 6`         |
| `**=`    | возведение в степень  | `x **= 7` | `x = x ** 7`        |

| Характеристика         | В математике           | В Python (программировании)          |
| ---------------------- | ---------------------- | ------------------------------------ |
| Назначение             | Проверка равенства     | Присваивание значения                |
| Работает двусторонне?  | Да (`x = 5` ⇄ `5 = x`) | Нет (всегда `переменная = значение`) |
| `x = x + 1` допустимо? | ❌ Нет (ошибка)         | ✅ Да (изменение значения)            |
| Тип операции           | Логическая             | Оперативная (выполняемая)            |

🧠 Подведём итог:

*    В математике `=` — это утверждение: «x равен 5».

*    В Пайтоне `=` — это команда: «сохрани 5 в переменную x».

*    В Пайтоне переменные живут во времени и могут изменяться.
